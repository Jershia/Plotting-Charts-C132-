```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Star_Data(output).csv")
df.head()
```

Convert to list

```python
mass=df["Mass"].tolist()
radius=df["Radius"].tolist()
gravity=df["Gravity"].tolist()
distance=df["Distance"].tolist()
```

Sort using sort method

```python
mass.sort()
radius.sort()
gravity.sort()
```

Chart of radius and mass

```python
plt.plot(radius,mass)
plt.title("Radius and Mass")
plt.xlabel("Radius")
plt.ylabel("Mass")
plt.show()
```

Chart of gravity and mass

```python
plt.plot(gravity,mass)
plt.title("Gravity and Mass")
plt.xlabel("Mass")
plt.ylabel("Gravity")
plt.show()
```
